### Problem 2.12
The goal of this problem is determine the frequency of a seismically-excited system with a response amplitude of 4mm to an input of 0.001sin(50t)m.

```python
# Problem 2.12 - Let's check the solution
t = np.linspace(0,1,1001)

# Set up the harmonic, seismic input
y_bar = 0.001                      # input amplitude
w = 50.                            # input frequency
y = y_bar * np.sin(w*t)            # The seismic (position) input

# Plug in the answer
wn = np.sqrt(10000.0/3)            # natural frequency (rad/s)

# x_bar is the amplitude of the vibration of x
x_bar = (wn**2)/(wn**2 - w**2) * y_bar
print 'The amplitude of vibration at 57.74 rad/s is ' + str(x_bar) + 'm.'

# generate the steady-state response
x = x_bar * np.sin(w*t)

# Make the figure pretty, then plot the results
#   "pretty" parameters selected based on pdf output, not screen output
#   Many of these setting could also be made default by the .matplotlibrc file
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

xlabel('Time (s)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel('Position (m)',family='serif',fontsize=22,weight='bold',labelpad=10)


plot(t, y, linestyle = '--', linewidth=2, label='$y(t)$')
plot(t, x, linewidth=2, label='$x(t)$')
ylim(-0.005, 0.006)

leg = legend(loc='upper right', ncol = 2, fancybox=True)
ltext  = leg.get_texts()
setp(ltext,fontsize=16)

# Adjust the page layout filling the page using the new tight_layout command
tight_layout(pad=0.5)

# Uncomment below to save the figure as a pdf
# savefig('Problem2p12.pdf')
```

### Problem 2.47
This problem asks to write:

$\quad x(t) = 4 \cos{3t} - 2\sin{3t} $

in complex form such that the real part gives the $x(t)$ above.

```python
# Problem 2.47
import sympy
sympy.init_printing()

t = sympy.symbols('t', real=True)

x = 4 * sympy.cos(3*t) - 2 * sympy.sin(3*t)

# Rewrite x(t) as an exponential form 

# First using the sympy rewrite function
x_exp = x.rewrite(exp)

# Then, our "hand-generated" solution
x_sol = (4 + 2*sympy.I) * sympy.exp(sympy.I*3*t)

# output the results beautifully
display(x_exp)
display(x_sol)
```

Notice that this is not the simplest form. It is using two trig identities:

$ \quad \cos{x} = \frac{e^{ix} + e^{-ix}}{2} $ and 

$ \quad \sin{x} = \frac{e^{ix} - e^{-ix}}{2i} $

```python
# Now take the real part of the answer to check - we should get back the original function
sympy.re( x_sol )
```

Let's check the answer numerically.

```python
time = np.linspace(0,10,100)

# The original function
x_num = 4 * np.cos(3*time) - 2 * np.sin(3*time)

# Our complex exponential form
x_sol_num = (4 + 2j) * np.exp(3j*time)
x_sol_num_real = np.real(x_sol_num)

# Make the figure pretty, then plot the results
#   "pretty" parameters selected based on pdf output, not screen output
#   Many of these setting could also be made default by the .matplotlibrc file
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

xlabel('Time (s)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel('$x(t)$',family='serif',fontsize=22,weight='bold',labelpad=10)


plot(time, x_num, linewidth=2, label='Orignial $x(t)$')
plot(time, x_sol_num_real, linestyle = '--', linewidth=2, label='Re$[x_{exp}(t)]$')
ylim(-5, 7)

leg = legend(loc='upper right', ncol = 2, fancybox=True)
ltext  = leg.get_texts()
setp(ltext,fontsize=16)

# Adjust the page layout filling the page using the new tight_layout command
tight_layout(pad=0.5)

# Uncomment below to save the figure as a pdf
# savefig('Problem2p47.pdf')
```
