### Problem 1.4
The goal of this problem is to represent $ 2 \cos{3t} + 4 \sin{2t} $ in exponential form, as a sum of $e^{3it}$ and $e^{-3it}$.

```python
# Problem 1.4 using SymPy - A symbolic math package
from sympy import *

def accel_piece1(t):
    accel = (0.5*t + 2)
    return accel

def accel_piece2(t):
    accel = 4
    return accel

x = Symbol('x')
t = Symbol('t')
c2 = Symbol('c2')

vel_piece1 = integrate(accel_piece1(x), x)
vel_piece2 = integrate(accel_piece2(x), x)    

# Now, evaluate the first piece to get the initial condition for the 2nd piece
vel_piece1_def = integrate(accel_piece1(x), (x,0,4))

print 'Velocity in Piece 1 (0 <= t <= 4):        ', + vel_piece1
print 'Velocity in Piece 2 (t > 4), no const:    ', + vel_piece2

# We need this to solve for the integration constant in piece2
print '\nVelocity at t = 4:                          ', + vel_piece1_def

def vel_piece1(t):
    vel = 0.25*t**2 + 2.0*t
    return vel

def vel_piece2(t,const):
    vel = 4*t + const
    return vel

# Add the constant for the initial condition
const2 = solve(vel_piece2(4.,c2) - vel_piece1_def, c2)

print '\nVelocity in Piece 2 (t > 4), with const:  ', + vel_piece2(x, const2[0])
```

```python
# Now let's integrate again, this time we can just do definite integrals
pos_piece1 = integrate(vel_piece1(x),(x,0,4))
pos_piece2 = integrate(vel_piece2(x, const2[0]), (x,4,8)) + pos_piece1

print 'The position after 8s is: {:.2f}.'.format(float(pos_piece2))
```

### Problem 1.30
This problem is looking at the linearization of a nonlinear spring at various points in its operating range.

```python
# Problem 1.30
from sympy import *
x = Symbol('x')

# Define the constants in the problem
m1 = 0.4209    # 1st mass in problem
m2 = 0.8418    # 2nd mass in problem
g = 9.81       # Gravity (m/(s**2))

# Define the spring force - include 100x conversion from N/cm to N/m
f_sp = (10*x - 4*x**2 + x**3)

# Solve for the m1 equilibrium position
x_eq1 = solve(f_sp - m1*g, x)

# Solve for the m2 equilibrium position
x_eq2 = solve(f_sp - m2*g, x)



# To linearize the spring, we want the slope of f_sp at each equilibrium point
f_sp_dot = diff(f_sp, x)

# Evaluate the derivative (the slope of f_sp) at the 1st equilibrium point
k_eq1 = f_sp_dot.subs(x, x_eq1[0])*100.

# Evaluate the derivative (the slope of f_sp) at the 2nd equilibrium point
k_eq2 = f_sp_dot.subs(x, x_eq2[0])*100.

# Get the two natural frequencies - We need to float to convert from 
#  the between the symbolic representation and numeric
wn1 = np.sqrt(float(k_eq1) / m1)
wn2 = np.sqrt(float(k_eq2) / m2)

# Uncomment below to print the answers
# print x_eq1
# print x_eq2
# print k_eq1
# print k_eq2
# print wn1
# print wn2

print 'There is a {:.2f}% difference between the natural frequencies.'.format((1 - wn2/wn1) * 100) 
```

Let's plot the linearizations along with the spring force. The linearized version are just the tangent of the spring force at the point of linearization.

```python
# Let's plot the spring force and linearized versions

def f_Nonlinear(spring_displacement):
    # Define the nonlinear force as N/m
    f_Nonlinear = 100*(10.*spring_displacement - 4.*spring_displacement**2 + spring_displacement**3)
    return f_Nonlinear

# Define an array of displacements for plotting
displacement = np.linspace(0,2,500)

# Fill in an array for the nonlinear force
f_Nonlin = f_Nonlinear(displacement)

# We need to include a y-intercept offset to plot the linear approx in the right place
f_sp1 = float(k_eq1) * displacement + (f_Nonlinear(x_eq1[0]) - float(k_eq1)*float(x_eq1[0]))
f_sp2 = float(k_eq2) * displacement + (f_Nonlinear(x_eq2[0]) - float(k_eq2)*float(x_eq2[0]))

# Make the figure pretty, then plot the results
#   "pretty" parameters selected based on pdf output, not screen output
#   Many of these setting could also be made default by the .matplotlibrc file
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

xlabel('Displacement (m)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel('Force (N)',family='serif',fontsize=22,weight='bold',labelpad=8)

ax.plot(displacement, f_Nonlin, color='blue', linewidth=2, label='Nonlinear')
ax.plot(displacement, f_sp1, 'r--', linewidth=2, label=r'$m$ = 0.4209kg')
ax.plot(displacement, f_sp2, 'k-.', linewidth=2, label=r'$m$ = 0.8418kg')
xlim(0.0,2.0)


leg = legend(loc='lower right', ncol = 1, fancybox=True)
ltext  = leg.get_texts()
setp(ltext,fontsize=16)

# Uncomment below to save the figure as a pdf
#savefig('Problem1p30_NonlinSpring.pdf',dpi=300)
```

### Problem 1.62
In this problem, we're looking at the system below in Figure 1. We're going to compare this system, whose pendulum bar has mass to the simple pendulum model (whose bar is massless).

```python
# Problem 1.62 - let's compare the massless-bar pendulum to the system in this problem
import control             # import the control system functions

l = 2.0                    # Pendulum Length (m)
m1 = 1.0                   # Point mass at end of bar (kg)
m2 = 1.0                   # Mass of bar itself (kg)

I_PointMass = m1*l**2      # Moment of Intertia of the point mass
I_Bar = 1.0/3.0 * l**2     # Moment of Intertia of the bar

g = 9.81                   # Gravity (m/s**2)


# Define the massless-bar pendulum system (often called a "simple" pendulum) 
num = [1.]
den = [1., 0., m1*g*l/I_PointMass]

sys_simple = control.tf(num,den)

# Define the time vector and an input of all zeros (we're only looking at initial condition response)
t = np.linspace(0,10,1000)
yd = np.zeros_like(t)

x0 = [0.,15*np.pi/180.]

# run the simulation for the simple pendulum
[T,yout_simple,x_out] = control.forced_response(sys_simple,t,yd,x0)


# Define the pendulum system from this problem
num = [1.]
den = [1., 0., (m1*g*l + m2*g*l/2.)/(I_PointMass+I_Bar)]

sys_mass = control.tf(num,den)

# run the simulation for the "bar has mass" case
[T,yout_mass,x_out] = control.forced_response(sys_mass,t,yd,x0)
```

```python
# Plotting for Problem 1.62

# Make the figure pretty, then plot the results
#   "pretty" parameters selected based on pdf output, not screen output
#   Many of these setting could also be made default by the .matplotlibrc file
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

xlabel('Time (s)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel('Angle (deg)',family='serif',fontsize=22,weight='bold',labelpad=10)

plot(T,yout_simple * 180./np.pi,color="blue",linewidth=2,label='Simple')
plot(T,yout_mass * 180./np.pi,color="red", linestyle='--',linewidth=2,label='Bar has Mass')
ylim(-15., 20)


leg = legend(loc='upper right', ncol = 2, fancybox=True)
ltext  = leg.get_texts() 
setp(ltext,family='serif',fontsize=16)
```
