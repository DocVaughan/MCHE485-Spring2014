## Problem 1
This is a simple mass-spring-damper problem.

** a. ** The equation of motion for the system is:

$ \quad m \ddot{x} + c \dot{x} + \frac{k}{2}x = \frac{k}{2}y $

We could also write this equation in terms of the damping ratio, $\zeta$, and natural frequency, $\omega_n$, by dividing by $m$.

$ \quad \ddot{x} + 2\zeta\omega_n\dot{x} + \omega_n^2x = \omega_n^2y$.

where

$ \quad \omega_n^2 = \frac{k}{2m} \quad$ and $ \quad 2 \zeta \omega_n = \frac{c}{m}$.    
<br>
**b.** So, the natural frequency is: 

$ \quad \omega_n = \sqrt{\frac{k}{2m}}$

<br>
**c.** See below for the response plots.

**d.** We can write the response as:

$ \quad x(t) = e^{-\zeta\omega_nt}\left(b_1 \cos{\omega_d t} + b_2 \sin{\omega_d t}\right) $

To use this equation, we need to solve for $b_1$ and $b_2$ using the initial conditions. Solving the equation for generic intial velocity, $\dot{x} = v_0$, and a generic initial displacement, $x = x_0$, we find:

$ \quad x(t) = e^{-\zeta\omega_nt}\left(x_0 \cos{\omega_d t} + \frac{\zeta \omega_n x_0 + v_0}{\omega_d} \sin{\omega_d t}\right) $

<br>
**e.** We can define the normalized transfer function as:

$ \quad G(\Omega) = \sqrt{\frac{1 + (2\zeta\Omega)^2}{(1-\Omega^2)^2 + (2\zeta\Omega)^2}} e^{i \phi} $

where

$ \quad \Omega = \frac{\omega}{\omega_n}$

and

$ \quad \phi = \tan^{-1}\left({2\zeta\Omega}\right) - \tan^{-1}\left({\frac{2\zeta\Omega}{1-\Omega^2}}\right)$

<br>
**f.** See the plots below.

$ \quad \theta(t) = \bar{\theta} e^{i\omega t}$

$ \quad \frac{\bar{\theta}}{\bar{\tau}}  = \frac{1}{I}\frac{1}{(\omega_n^2 - \omega^2) + (2i\zeta\omega\omega_n)}$

```python
t = np.linspace(0,8,8000)             # define the time to look at, easiest to just choose 1 period
w0 = 2.0*np.pi/t[-1]                  # define the fundamental frequency (here, I know t(end)=tau)
tau_0 = 2.0*np.pi/w0                  # define fundamental period based on w0

F0 = 1.0
y = zeros((len(t),))

for ii in xrange(len(t)):
    if t[ii] <= 2.0:
        y[ii] = 2.0/4.0 * F0 * t[ii]
        
    elif t[ii] > 2.0 and t[ii] <= 4:
        y[ii] = -2.0/4.0 * F0 * t[ii] + 2*F0
        
    elif t[ii] > 4.0 and t[ii] <= 6:
        y[ii] = 2.0/4.0 * F0 * t[ii] - 2*F0
        
    elif t[ii] > 6.0:
        y[ii] = -2.0/4.0 * F0 * t[ii] + 4*F0
        
```

```python
#--------- This is the actual Numerical Fourier Expansion --------
#
# NOTE: SciPy trapz command is trapz(y,t) (opposite order of MATLAB/Octave)

# define the number of terms to use in the approximation
num_terms = 7

# get the a0 term
a0 = w0/(2.0*np.pi)*np.trapz(y,t)  

# fill arrays with zeros - Good practice, it speeds computation in large problems
a = np.zeros((num_terms,))
b = np.zeros((num_terms,))
integral_cos = np.zeros((len(t),num_terms))
integral_sin = np.zeros((len(t),num_terms))
sin_term = np.zeros((num_terms,len(t)))
cos_term = np.zeros((num_terms,len(t)))

# cycle through the 1 to num_terms Fourier coefficients (a_n and b_n)
for n in range(num_terms):

    # a_n calculations
    integral_cos[:,n] = y * np.cos((n+1)*w0*t)         # define the integral "interior"
    a[n] = w0/np.pi * np.trapz(integral_cos[:,n],t)    # solve for a_n

    # b_n calculations
    integral_sin[:,n] = y * np.sin((n+1)*w0*t)         # define the integral "interior"
    b[n] = w0/np.pi * np.trapz(integral_sin[:,n],t)    # solve for b_n
    
    sin_term[n,:] = np.sin((n+1)*w0*t)                 # calculate the nth sine term
    cos_term[n,:] = np.cos((n+1)*w0*t)                 # calculate the nth cosine_term


# Generate the approximate input based on the Fourier coeff. calculated above
approx = np.zeros_like(t) #First fill with zeros

for ii in range(len(t)):
     approx[ii] = a0 + np.sum(a * cos_term[:,ii],0) + np.sum(b * sin_term[:,ii],0)
```

## Problem 4
The system in this problem has both a viscous damper and friction.

**a.** It's equation of motion is:

$ \quad m \ddot{x} + \mu N sgn(\dot{x}) + k_{eq} x = 0 $

where $k_{eq}$ represents the equivalent spring cosntant for the three-spring set:

$ \quad k_{eq} = \frac{2 k_1 k_2}{2k_1 + k_2} $

We could also divided by $m$ to write this equation in terms of the natural frequency $\omega_n = \sqrt{k_{eq}/m}$.

$ \quad \ddot{x} + \frac{\mu N}{m} sgn(\dot{x}) + \omega_n^2 x = 0 $

## Problem 5

** a.**
$\quad \begin{bmatrix}m_1 & 0 \\ 0 & m_2\end{bmatrix}\begin{bmatrix}\ddot{x}_1 \\ \ddot{x}_2\end{bmatrix} + \begin{bmatrix}\phantom{-}k & -k \\ -k & \hphantom{-}k\end{bmatrix}\begin{bmatrix}x_1 \\ x_2\end{bmatrix} = \begin{bmatrix}f \\ 0\end{bmatrix} $

```python
# Define the matrices
m1 = 1.0
m2 = 1.0

k = 2.0*np.pi**2

M = np.asarray([[m1, 0],
                [0,  m2]])

K = np.asarray([[k,  -k],
                [-k,  k]])
```

```python
# We'll use the scipy version of the linear algebra
from scipy import linalg

eigenvals, eigenvects = linalg.eigh(K,M)
```

**b.** These eigenvectors are in normalized form. We can write them otherwise, as long as we maintain the relationship between the coordinates. So the eigenvalues and eignevectors are:

$ \quad \omega_1^2 = 0.0,\ X_1 = \begin{bmatrix}1 \\ 1\end{bmatrix} \quad \mbox{and} \quad \omega_2^2 = 39.48, \ X_2 = \begin{bmatrix}\phantom{-}1 \\ -1\end{bmatrix} $

**c.** **i.** 
See below for the plot of the response to:

$ \quad x_1(0) = x_0, \ x_2(0) = x_0, \ \dot{x}_1(0) = \dot{x}_2(0) = 0 $

```python
# Define the equations of motion

# Define the system as a series of 1st order ODEs (beginnings of state-space form)
def Prob4_eq_of_motion(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x, x_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m1, m2, k]
    """
    x1, x1_dot, x2, x2_dot = w
    m1, m2, k = p

    # Create sysODE = (x', x_dot'): - Here, we're assuming f(t) = 0
    sysODE = [x1_dot,
             (-k*x1 + k*x2) / m1,
             x2_dot,
             (k*x1 - k*x2) / m2]
    return sysODE
```

```python
from scipy.integrate import odeint  

# Set up simulation parameters 

# ODE solver parameters
abserr = 1.0e-9
relerr = 1.0e-9
max_step = 0.01
stoptime = 3.0
numpoints = 10001

# Create the time samples for the output of the ODE solver
t = np.linspace(0.,stoptime,numpoints)

# Initial conditions
x1_init = 0.5                       # initial x1 position
x1_dot_init = 0.0                    # initial x1 velocity
x2_init = 0.5                       # initial x2 position
x2_dot_init = 0.0                    # initial x2 velocity

# Pack the parameters and initial conditions into arrays 
p = [m1, m2, k]
x0 = [x1_init, x1_dot_init, x2_init, x2_dot_init]

# Call the ODE solver.
resp_ci = odeint(Prob4_eq_of_motion, x0, t, args=(p,), atol=abserr, rtol=relerr,  hmax=max_step)
```

where

```python
F1 = 1.0
F2 = 0.0

F = [F1,F2]

w = np.linspace(0.1,10,1200)
X = np.zeros((len(w),2))

# This is (K-w^2 M)^-1 * F
for ii, freq in enumerate(w):
    X[ii,:] = np.dot(linalg.inv(K - freq**2 * M), F)

# Let's mask the discontinuity, so it isn't plotted
pos = np.where(np.abs(X[:,0]) >= 25)
X[pos,:] = np.nan
w[pos] = np.nan
```

To be normal, both of the following equations needs to be true.

$ \quad X_1^T M X_1 = 1 $ 

and 

$ \quad X_2^T M X_2 = 1 $.

We could also form 

$ \quad U = \begin{bmatrix}X_1 & X_2\end{bmatrix}$

where $X_1$ and $X_2$ are the eigenvectors as columns to check normality in matrix form. In this form,

$ \quad U^T M U = I$

where $I$ is the identity matrix, would indicate normality.

```python
U = np.array([[1, 1],[1,-1]])

# Write the matrix form of the normalization check.
np.dot(np.dot(U.T,M),U)
```

To normalize, we simpliy divide each eigenvector by the sqaure root of result of $ X_i^T M X_i$.

$ \quad \tilde{X}_i = \frac{1}{\sqrt{X_i^T M X_i}}X_i$

```python
# For the first eigenvector:
X1 = np.array([[1],[1]])

X1_norm = 1.0/np.sqrt(np.dot(np.dot(X1.T,M),X1))*X1

print 'The normalized form of \n'
print np.array_str(X1,precision=4,suppress_small=True) 
print '\n is \n' 
print np.array_str(X1_norm,precision=4,suppress_small=True)
```
